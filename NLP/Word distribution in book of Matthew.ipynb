## Read file

```python
doc = open('matthew.txt') # Create handle
doctext = doc.read()
```

## Clean text file

```python
from string import punctuation # import the punctuation library
txt = ''.join(c for c in doctext if not c.isdigit()) # Get rid of digits
txt = ''.join(c for c in txt if c not in punctuation) # get rid of punctuations
txt = ''.join(c for c in txt if c not in ['"',')','(','!','?','”','“','‘','——', '—']) # eliminate any remaining signs and symbols
old = ['dI','eJ','iP','jH','lN','nB','vR','yJ','zJ','cA','eM','tT','cT','fD','bI','lA','kF','mB','mT','jC','nA'] # links to index
new = ['I','J','P','H','N','B','R','J','J','A','M','T','T','D','I','A','F','B','T','C','A'] # replacements
for n in range(len(old)):
    matttxt = txt.replace(old[n], new[n]).lower() # eliminate the links
```

```python
print(matttxt)
```

## Get word count

```python
import pandas as pd
import nltk
from nltk.probability import FreqDist
nltk.download('punkt') 

# Tokenize the text into induividual words
words = nltk.tokenize.word_tokenize(matttxt)

# Get frequency distribution of words into a dataframe
freqdis = FreqDist(words)
count_frame = pd.DataFrame(freqdis, index =[0]).T
count_frame.columns = ['Count']
print(count_frame)
```

## Plot distribution as a pareto chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
counts = count_frame.sort_values('Count', ascending = False) # sort the count of words in descending order

# Display top 60 words as a bar chart
fig = plt.figure(figsize = (16,6))
ax = fig.gca()
counts['Count'][:60].plot(kind = 'bar', color = 'm', ax = ax)
plt.show()
```

## Remove stopwords

```python
# Get standard stowords from NLTK
nltk.download('stopwords')
from nltk.corpus import stopwords 
```

```python
# Filter out the stopwords
matttxt = ' '.join([word for word in matttxt.split() if word not in (stopwords.words('english'))])
```

```python
print(matttxt)
```

## Get new word count

```python
# Tokenize words
words = nltk.tokenize.word_tokenize(matttxt)

# Get frequency distribution of words into a dataframe
from nltk.probability import FreqDist
freqdis = FreqDist(words)
count_frame = pd.DataFrame(freqdis, index = [0]).T
count_frame.columns = ['Counts']
```

## Plot frequency distribution of top 60 words in Matthew


```python
%matplotlib inline
counts = count_frame.sort_values('Counts', ascending = False)
fig = plt.figure(figsize = (16, 6))
ax = fig.gca()

counts['Counts'][:60].plot(kind = 'bar', ax = ax, color = 'r')
plt.xlabel('words')
plt.ylabel('Frequency')
plt.title('Word Distribution in Book of Mathew')
plt.subplots_adjust(bottom = 0.2)
plt.show()
```

The corpus for this project is taken from the 2013 edition of "[New World Translation of the Holy Scriptures](https://www.jw.org/en/publications/bible/study-bible/books/matthew/1/)"
