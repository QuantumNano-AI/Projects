## Get corpus

```python
doc = open('Luke.txt')
luktxt = doc.read()
```

## Clean

```python
from string import punctuation
txt = ''.join(c for c in luktxt if not c.isdigit())
txt = ''.join(c for c in txt if c not in punctuation)
luktxt = ''.join(c for c in txt if c not in ['”','“','’','‘', '——', '—']).lower()
```

```python
# print(luktxt)
```

# Remove Stopwords

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

luktxt = ' '.join(c for c in luktxt.split() if c not in stopwords.words('english'))
```

# Get word count

```python
import nltk
import pandas as pd
import numpy as np
nltk.download('punkt')
from nltk.probability import FreqDist

#Tokenize each individual word and make into a DataFrame
words = nltk.tokenize.word_tokenize(luktxt)
freqdis = FreqDist(words)
count_frame = pd.DataFrame(freqdis, index =[0]).T
count_frame.columns = ['Counts']
```

## Plot Word distribution

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (16, 6))
ax = fig.gca()

# Sort words in order of frequency
counts = count_frame.sort_values('Counts', ascending = False)
counts.Counts[:60].plot(kind = 'bar', ax = ax, color = 'r')
ax.set_title('Word distribution for Luke')
ax.set_ylabel('Frequency')
ax.set_xlabel('Words')
plt.show()
```

The corpus for this project is taken from the 2013 edition of "[New World Translation of the Holy Scriptures](https://www.jw.org/en/publications/bible/study-bible/books/matthew/1/)"
